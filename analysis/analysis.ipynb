```python
def exactly_one(cond, it):
    ts = [x for x in it if cond(x)]
    if len(ts) == 1:
        return ts[0]
    else:
        return None
```

```python
f2 = set()
for f in faces3d:
    if len(f.ins) == 2:
        w1 = lambda x: x.p1.level == 2 and not x.p2.level == 2
        w2 = lambda x: not x.p1.level == 2 and x.p2.level == 2
        a1 = exactly_one(w1, f.ins)
        a2 = exactly_one(w2, f.ins)
        
        if a1 and a2 and a1 != a2:
            f2.add(f)
```

```python
def is_prod_of_two_arrows(f):
    return f.p1.level == 1 and f.p2.level == 1
```

```python
f3 = set()
for f in faces3d:
    if len(f.ins) == 3:
        w1 = lambda x: x.p1.level == 2 and not x.p2.level == 2
        w2 = lambda x: not x.p1.level == 2 and x.p2.level == 2
        w3 = is_prod_of_two_arrows
        a1 = exactly_one(w1, f.ins)
        a2 = exactly_one(w2, f.ins)
        a3 = exactly_one(w3, f.ins)
        
        if a1 and a2 and a1 != a2 and a3:
            f3.add(f)
```

```python
f4 = set()
for f in faces3d:
    w1 = lambda x: x.p1.level == 2 and x.p2.level == 2
    a1 = exactly_one(w1, f.ins)
    a2 = any(is_prod_of_two_arrows(x) for x in f.ins)
    if a1 and a2:
        f4.add(f)
```

```python
l2_114 = [f for f in f3 if len(f.ins) == 3]

def select_domain_face(NUMBER):
    return [x for x in l2_114[NUMBER].ins]
```

```python
out = []

def t_str(x):
    return "\n".join([str((t.p1, t.p2)) for t in x.ins])

for i in range(len(l2_114)):
    fs = select_domain_face(i)
    fs = sorted([t_str(f) for f in fs], key=lambda x: len(x))
    
    out.append("Number: {}".format(i))
    for f in fs:
        out.append(f)
        out.append("")
    out.append("")
out = [o + "\n" for o in out]
```
